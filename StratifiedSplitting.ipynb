```python
import pandas as pd
df = pd.read_csv('Bent_AllDesc_Stratified.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
for i in range(100):
    strat_train_set, strat_test_set = train_test_split(df, test_size=.25, random_state=i, stratify=df['Group'])
    train_string = "trainset{}.csv".format(i)
    test_string = "testset{}.csv".format(i)
    strat_train_set.to_csv(train_string)
    strat_test_set.to_csv(test_string)
```
